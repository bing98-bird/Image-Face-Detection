**Face detection**

```python
face_cascade = cv2.CascadeClassifier("DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):

  face_img = img.copy()

  face_rects = face_cascade.detectMultiScale(face_img)

  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

  return face_img
```

**Face detection on Einstein**

```python
result = detect_face(einstein)
plt.imshow(result,cmap='gray')
```

*Gaussian Blur Technique on Einstein*

```python
gaussian_einstein = cv2.GaussianBlur(einstein, (155, 155), 0)
plt.imshow(cv2.cvtColor(gaussian_einstein, cv2.COLOR_BGR2RGB))
plt.title('Increased Gaussian Blur Applied')
plt.show()
```

```python
result = detect_face(gaussian_einstein)
plt.imshow(result,cmap='gray')
```

*Median Blur Technique on Einstein*

```python
median_einstein = cv2.medianBlur(einstein, 85)
plt.imshow(cv2.cvtColor(median_einstein, cv2.COLOR_BGR2RGB))
plt.title('Increased Median Blur Applied')
plt.show()
```

```python
result = detect_face(median_einstein)
plt.imshow(result,cmap='gray')
```

*Average Blur Technqiue on Einstein*

```python
blur_einstein = cv2.blur(einstein, (71, 71))
plt.imshow(cv2.cvtColor(blur_einstein, cv2.COLOR_BGR2RGB))
plt.title('Increased blur Blur Applied')
plt.show()
```

```python
result = detect_face(blur_einstein)
plt.imshow(result,cmap='gray')
```

*Bilateral Blur Technique on Einstein*

```python
bilateral_einstein = cv2.bilateralFilter(einstein, 100, 600, 600)
plt.imshow(cv2.cvtColor(bilateral_einstein, cv2.COLOR_BGR2RGB))
plt.title('Further Increased Bilateral Filter Applied')
plt.show()
```

```python
result = detect_face(bilateral_einstein)
plt.imshow(result,cmap='gray')
```

**Adjusted Face Detection Funtion**

```python
def adj_detect_face(img):

  face_img = img.copy()

  face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.3,minNeighbors=5)

  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

  return face_img
```

```python
result = adj_detect_face(einstein)
plt.imshow(result,cmap='gray')
```

```python
result = adj_detect_face(wife)
plt.imshow(result,cmap='gray')
```

**Eye detection**

```python
eye_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):

  face_img = img.copy()

  eyes_rects = eye_cascade.detectMultiScale(face_img,scaleFactor=1.32,minNeighbors=5)

  for (x,y,w,h) in eyes_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

  return face_img
```

```python
result = detect_eyes(wife)
plt.imshow(result,cmap='gray')
```
